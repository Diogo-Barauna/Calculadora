```python
import pandas as pd

data = pd.read_excel('maiores_empresas.xlsx')

data.head(10)
```

```python
prop = {'train': 0.85, 'test': 0.15}

data_train = data.sample(frac=prop['train'])
data_test = data.drop(data_train.index)


print("Tamanho da base de treino: ", len(data_train))
print("Tamanho da base de teste: ", len(data_test))

data_train.head(10)
```

```python
from matplotlib import pyplot as plt

bancario = data_train[data_train['Setor'] == 'Bancário']
comercio_atacadista = data_train[data_train['Setor'] == 'Comércio Atacadista']
comercio_varejista = data_train[data_train['Setor'] == 'Comércio Varejista']
construcao = data_train[data_train['Setor'] == 'Construção']
imobiliario = data_train[data_train['Setor'] == 'Imobiliário']
industrias_transformacao = data_train[data_train['Setor'] == 'Indústrias da transformação']
outros = data_train[data_train['Setor'] == 'Outros']
produtos_farmaceuticos = data_train[data_train['Setor'] == 'Produtos farmacêuticos']
servicos = data_train[data_train['Setor'] == 'Serviços']

plt.scatter(bancario['rssiOne'], bancario['rssiTwo'], label='Bancário')
plt.scatter(comercio_atacadista['rssiOne'], comercio_atacadista['rssiTwo'], label='Comércio Atacadista')
plt.scatter(comercio_varejista['rssiOne'], comercio_varejista['rssiTwo'], label='Comércio Varejista')
plt.scatter(construcao['rssiOne'], construcao['rssiTwo'], label='Construção')
plt.scatter(imobiliario['rssiOne'], imobiliario['rssiTwo'], label='Imobiliário')
plt.scatter(industrias_transformacao['rssiOne'], industrias_transformacao['rssiTwo'], label='Indústrias da Transformação')
plt.scatter(outros['rssiOne'], outros['rssiTwo'], label='Outros')
plt.scatter(produtos_farmaceuticos['rssiOne'], produtos_farmaceuticos['rssiTwo'], label='Produtos Farmacêuticos')
plt.scatter(servicos['rssiOne'], servicos['rssiTwo'], label='Serviços')

plt.legend()
plt.xlabel('RSSI One')
plt.ylabel('RSSI Two')
plt.title('Gráfico de Dispersão por Setor')
plt.show()
```
